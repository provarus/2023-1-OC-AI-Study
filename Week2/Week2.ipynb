```python
import matplotlib.pyplot as plt
import seaborn as sns

penguin_df = sns.load_dataset("penguins")

sns.scatterplot(x="body_mass_g", y="bill_length_mm", data=penguin_df)

plt.title('Body Mass vs Bill Length')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [123, 1, 59, 40, 10, 100]

plt.figure(figsize=(12,5))

plt.subplot(1,3,1)
plt.plot(x, y)
plt.title('Line Plot')

plt.subplot(1,3,2)
plt.bar(x, y)
plt.title('Bar Plot')

plt.subplot(1,3,3)
plt.scatter(x, y)
plt.title('Scatter Plot')

plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

penguin_df = sns.load_dataset("penguins")

sns.kdeplot(data=penguin_df, x="body_mass_g", color="orange", fill=True)

plt.title('Distribution of Penguin Body Mass')
plt.ylabel('Density')

plt.show()
```

```python
import matplotlib.pyplot as plt

cars = ['AUDI', 'BMW', 'FORD', 'TESLA', 'JAGUAR']
data = [23, 29, 41, 12, 17]

explode = (0, 0, 0.2, 0, 0)

plt.pie(data, labels=cars, explode=explode, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Car Brands')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')

fig, ax = plt.subplots(figsize=(12, 6))

sns.boxplot(x='size', y='tip', data=tips, ax=ax)


ax.set_title('Distribution of Tips by Party Size')

plt.show()
```

```python
import matplotlib.pyplot as plt

X = ["Tom", "Amy", "Jim", "Bill", "Pegi"]
Y = [8, 23, 52, 37, 64]

fig, ax = plt.subplots(figsize=(8,6))
ax.bar(X, Y, color='lightgray', edgecolor='gray')
ax.set_xlabel('Name', fontsize=14, fontweight='bold', color='gray', labelpad=12)
ax.set_ylabel('Score', fontsize=14, fontweight='bold', color='gray', labelpad=12)
plt.show()
```

```python
import matplotlib.pyplot as plt

X = ["Tom", "Amy", "Jim", "Bill", "Pegi"]
Y = [8, 23, 52, 37, 64]

fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(X, Y, color="lightgray", edgecolor="gray")
ax.set_xlabel("Name", fontsize=14, fontweight="bold", color="gray", labelpad=12)
ax.set_ylabel("Value", fontsize=14, fontweight="bold", color="gray", labelpad=12)
ax.tick_params(axis="x", labelrotation=45)
ax.axhline(y=45, color="blue", linestyle="--")
ax.axhline(y=15, color="red", linestyle="--")
ax.axhspan(0, 15, facecolor="mistyrose", alpha=0.5)
for i, v in enumerate(Y):
    ax.text(i, v-4, str(v), ha="center", color="black", fontweight="bold")
plt.show()
```
